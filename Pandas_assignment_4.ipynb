3.Detect outliers in the Units_Sold column using the IQR method. What are the lower
and upper bounds?

```python
Q1=df['Units_Sold'].quantile(0.25)
print("Q1 for units_sold:",Q1)
Q3=df['Units_Sold'].quantile(0.75)
print("Q3 for units_sold:",Q3)
```

```python
IQR=Q3-Q1
print("IQR for units_sold:",IQR)
```

```python
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
```

```python
print("Lower bound:",lower_bound)
print("Upper bound:",upper_bound)
```
