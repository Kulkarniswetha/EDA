**SPECIES DISTRIBUTION**

3.How many samples belong to each species in the dataset?create a count of each species.

```python
species_counts=data['Species'].value_counts()
print("The count of each species in the dataset are:\n",species_counts)
```

**FEATURE RELATIONSHIPS**

4.what is the average sepal length for each species? similarly,what is the average petal width for each species?

```python
avg_sepal_length=data["SepalLengthCm"].mean()
print("The average sepal length for each species is:",avg_sepal_length)
```

```python
avg_petal_width=data["PetalWidthCm"].mean()
print("The average petal width for each species is:",avg_petal_width)
```

**FILTERING DATA**

5.Create a subset of the dataset containing only the samples where the petal length is greater than 1.5.How many such samples exist,and which species do they belong to?

```python
subset=data[data["PetalLengthCm"]>1.5]
print("The subset of the dataset containing only the samples where the petal length is greater than 1.5 is:\n",subset)
```

```python
#number of samples in the subset
num_samples=len(subset)
print("The number of samples in the subset is:",num_samples)
```

```python
#species counts in the subset
species_counts=subset["Species"].value_counts()
print("The species counts in the subset are:\n",species_counts)
```
