```python
import pandas as pd
```

```python
df2=pd.read_csv("/content/departments (2).csv")
```

```python
df1=pd.read_csv("/content/employees.csv")
```

```python
df1
```

```python
df2
```

```python
inner_join=pd.merge(df1,df2,on="Department_ID",how="inner")
inner_join
```

```python
left_join=pd.merge(df1,df2,on="Department_ID",how="left",suffixes=(" "," "))
left_join
```

```python
right_join=pd.merge(df1,df2,on="Department_ID",how="right")
right_join
```

```python
outer_join=pd.merge(df1,df2,on="Department_ID",how="outer")
outer_join
```

```python
Sum=df1.groupby("Department_ID")["Salary"].agg(['sum'])
Sum
```

```python
Average=df1.groupby("Department_ID")["Salary"].agg(["mean"])
Average
```

```python
joined_data= pd.merge(df1, df2, on='Department_ID', how='inner')
grouped_by_location = joined_data.groupby('Location').agg(
    num_employees=('Employee_ID', 'count'),
    total_salary=('Salary', 'sum')).reset_index()
print(grouped_by_location)
```

```python
filtered_employees = df1[df1['Salary'] > 50000]
avg_salary_filtered = filtered_employees.groupby('Department_ID')['Salary'].mean().reset_index()
print(" Employees (Salary > 50,000) - Average Salary by Department:\n", avg_salary_filtered)
```
